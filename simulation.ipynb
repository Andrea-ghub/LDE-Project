```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import sys
from utils import SIR_net, SIR_net_adaptive
```

```python
N = 2000 #number of nodes
ave_degree = 15
p = ave_degree / N  #edge probability
ERG = nx.erdos_renyi_graph(N, p)
nx.draw(ERG)
BAG = nx.barabasi_albert_graph(N, ave_degree)
#nx.draw(BAG)
print(len(nx.edges(ERG)))
print(len(nx.edges(BAG)))
```

```python
print('first simulation')
sim1 = SIR_net(ERG, 0.15, 0.1, initial_infecteds=np.arange(100))
print('second simulation')
sim2 = SIR_net_adaptive(ERG, BAG, beta=0.15, mu=0.1, r=0.99, pro=0.3, pol=0.05, p_sym=0.001, initial_infecteds=np.arange(100), initial_no_vax=np.arange(200,400))
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(10,6))
ax.plot(sim1[0], sim1[2], lw=3, c='red', label= "with No-Vax", alpha=0.6)
ax.plot(sim2[0], sim2[2], lw=3, c='springgreen', label= "SIR", alpha=0.6)
ax.legend(fontsize=15)
ax.set_title("Rewiring + No-Vax vs SIR", size=15, fontweight='bold')
ax.set_ylabel("Prevalence", size=15)
ax.set_xlabel("time", size=15)
```
